```python
import pandas as pd

# Read the sales_data_2.csv file into a DataFrame
sales_data = pd.read_csv('sales_data_2.csv')
# Convert the 'Invoice Date' column to a datetime object
sales_data['Date'] = pd.to_datetime(sales_data['Date'])
# Calculate the daily gross profit
sales_data['Daily Gross Profit'] = sales_data['Selling price'] - sales_data['Buying price']
# Calculate the 3-day average of the daily gross profit
sales_data['3-Day Average Daily Gross Profit'] = sales_data['Daily Gross Profit'].rolling(window=3).mean()
```

```python
import matplotlib.pyplot as plt

# Visualize the 3-day average daily gross profit
plt.plot(sales_data['Date'], sales_data['3-Day Average Daily Gross Profit'])
plt.xlabel('Date')
plt.ylabel('3-Day Average Daily Gross Profit')
plt.title('3-Day Average Daily Gross Profit Trend')
plt.show()
```
